## Importar Bibliotecas y Definir Funciones Auxiliares[cite: 1]


Importaremos las siguientes bibliotecas al laboratorio[cite: 1]


```python
# Requests nos permite hacer solicitudes HTTP que usaremos para obtener datos de una API[cite: 1]
import requests
# Pandas es una biblioteca de software escrita para el lenguaje de programación Python para manipulación y análisis de datos.[cite: 1]
import pandas as pd
# NumPy es una biblioteca para el lenguaje de programación Python, que agrega soporte para grandes arreglos y matrices multidimensionales, junto con una gran colección de funciones matemáticas de alto nivel para operar en estos arreglos.[cite: 1]
import numpy as np
# Datetime es una biblioteca que nos permite representar fechas[cite: 1]
import datetime

# Configurar esta opción imprimirá todas las columnas de un dataframe[cite: 1]
pd.set_option('display.max_columns', None)
# Configurar esta opción imprimirá todos los datos en una característica[cite: 1]
pd.set_option('display.max_colwidth', None)
```

A continuación definiremos una serie de funciones auxiliares que nos ayudarán a usar la API para extraer información usando números de identificación en los datos de lanzamiento.[cite: 1]

De la columna <code>rocket</code> nos gustaría conocer el nombre del propulsor (booster).[cite: 1]


```python
# Toma el conjunto de datos y usa la columna rocket para llamar a la API y agregar los datos a la lista[cite: 1]
def getBoosterVersion(data):
    for x in data['rocket']:
       if x:
        response = requests.get("https://api.spacexdata.com/v4/rockets/"+str(x)).json()
        BoosterVersion.append(response['name'])
```

Del <code>launchpad</code> nos gustaría saber el nombre del sitio de lanzamiento que se está utilizando, la longitud y la latitud.[cite: 1]


```python
# Toma el conjunto de datos y usa la columna launchpad para llamar a la API y agregar los datos a la lista[cite: 1]
def getLaunchSite(data):
    for x in data['launchpad']:
       if x:
         response = requests.get("https://api.spacexdata.com/v4/launchpads/"+str(x)).json()
         Longitude.append(response['longitude'])
         Latitude.append(response['latitude'])
         LaunchSite.append(response['name'])
```

De la <code>payload</code> (carga útil) nos gustaría conocer la masa de la carga útil y la órbita a la que se dirige.[cite: 1]


```python
# Toma el conjunto de datos y usa la columna payloads para llamar a la API y agregar los datos a las listas[cite: 1]
def getPayloadData(data):
    for load in data['payloads']:
       if load:
        response = requests.get("https://api.spacexdata.com/v4/payloads/"+load).json()
        PayloadMass.append(response['mass_kg'])
        Orbit.append(response['orbit'])
```

De <code>cores</code> nos gustaría conocer el resultado del aterrizaje, el tipo de aterrizaje, el número de vuelos con ese núcleo (core), si se usaron aletas de rejilla (gridfins), si el núcleo es reutilizado, si se usaron patas, la plataforma de aterrizaje utilizada, el bloque del núcleo que es un número utilizado para separar versiones de núcleos, el número de veces que este núcleo específico ha sido reutilizado, y el número de serie del núcleo.[cite: 1]


```python
# Toma el conjunto de datos y usa la columna cores para llamar a la API y agregar los datos a las listas[cite: 1]
def getCoreData(data):
    for core in data['cores']:
            if core['core'] != None:
                response = requests.get("https://api.spacexdata.com/v4/cores/"+core['core']).json()
                Block.append(response['block'])
                ReusedCount.append(response['reuse_count'])
                Serial.append(response['serial'])
            else:
                Block.append(None)
                ReusedCount.append(None)
                Serial.append(None)
            Outcome.append(str(core['landing_success'])+' '+str(core['landing_type']))
            Flights.append(core['flight'])
            GridFins.append(core['gridfins'])
            Reused.append(core['reused'])
            Legs.append(core['legs'])
            LandingPad.append(core['landpad'])
```

Ahora comencemos a solicitar datos de lanzamiento de cohetes a la API de SpaceX con la siguiente URL:[cite: 1]


```python
spacex_url="https://api.spacexdata.com/v4/launches/past"
```

```python
response = requests.get(spacex_url)
```

Comprobar el contenido de la respuesta[cite: 1]


```python
print(response.content)
```

Debería ver que la respuesta contiene una gran cantidad de información sobre los lanzamientos de SpaceX. A continuación, intentaremos descubrir información más relevante para este proyecto.[cite: 1]


### Tarea 1: Solicitar y analizar los datos de lanzamiento de SpaceX usando la solicitud GET[cite: 1]


Para hacer que los resultados JSON solicitados sean más consistentes, usaremos el siguiente objeto de respuesta estático para este proyecto:[cite: 1]

Ahora decodificamos el contenido de la respuesta como un JSON usando <code>.json()</code> y lo convertimos en un dataframe de Pandas usando <code>.json_normalize()</code>[cite: 1]


```python
# Use el método json_normalize para convertir el resultado json en un dataframe[cite: 1]
data = pd.json_normalize(response.json())
```

Usando el dataframe <code>data</code> imprima las primeras 5 filas[cite: 1]


```python
# Obtenga el encabezado (head) del dataframe[cite: 1]
data.head()
```

Notará que muchos de los datos son identificadores (IDs). Por ejemplo, la columna rocket no tiene información sobre el cohete, solo un número de identificación.[cite: 1]

Ahora usaremos la API nuevamente para obtener información sobre los lanzamientos utilizando los IDs dados para cada lanzamiento. Específicamente, usaremos las columnas <code>rocket</code>, <code>payloads</code>, <code>launchpad</code> y <code>cores</code>.[cite: 1]


```python
# Tomemos un subconjunto de nuestro dataframe manteniendo solo las características que queremos, el número de vuelo y date_utc.[cite: 1]
data = data[['rocket', 'payloads', 'launchpad', 'cores', 'flight_number', 'date_utc']]

# Eliminaremos filas con múltiples núcleos porque esos son cohetes falcon con 2 propulsores adicionales y filas que tienen múltiples cargas útiles en un solo cohete.[cite: 1]
data = data[data['cores'].map(len)==1]
data = data[data['payloads'].map(len)==1]

# Dado que payloads y cores son listas de tamaño 1, también extraeremos el valor único en la lista y reemplazaremos la característica.[cite: 1]
data['cores'] = data['cores'].map(lambda x : x[0])
data['payloads'] = data['payloads'].map(lambda x : x[0])

# También queremos convertir date_utc a un tipo de dato datetime y luego extraer la fecha dejando la hora[cite: 1]
data['date'] = pd.to_datetime(data['date_utc']).dt.date

# Usando la fecha restringiremos las fechas de los lanzamientos[cite: 1]
data = data[data['date'] <= datetime.date(2020, 11, 13)]
```

* Del <code>launchpad</code> nos gustaría saber el nombre del sitio de lanzamiento que se está utilizando, la longitud y la latitud.[cite: 1]

* **De <code>cores</code> nos gustaría conocer el resultado del aterrizaje, el tipo de aterrizaje, el número de vuelos con ese núcleo, si se usaron aletas de rejilla, si el núcleo es reutilizado, si se usaron patas, la plataforma de aterrizaje utilizada, el bloque del núcleo que es un número utilizado para separar versiones de núcleos, el número de veces que este núcleo específico ha sido reutilizado, y el número de serie del núcleo.**[cite: 1]

### Tarea 3: Tratar los Valores Faltantes[cite: 1]


Calcule a continuación la media para <code>PayloadMass</code> usando <code>.mean()</code>. Luego use la media y la función <code>.replace()</code> para reemplazar los valores `np.nan` en los datos con la media que calculó.[cite: 1]
