## Tareas

Ahora escriba y ejecute consultas SQL para resolver las tareas de la asignación.[cite: 3]

**Nota: Si los nombres de las columnas están en mayúsculas y minúsculas combinadas, enciérrelos entre comillas dobles
   Por ejemplo "Landing_Outcome"**[cite: 3]

### Tarea 1

##### Muestre los nombres de los sitios de lanzamiento únicos en la misión espacial[cite: 3]


```python
%sql SELECT DISTINCT Launch_Site FROM SPACEXTABLE;
```


### Tarea 2

#####  Muestre 5 registros donde los sitios de lanzamiento comiencen con la cadena 'CCA'[cite: 3]


```python
%sql SELECT * FROM SPACEXTABLE WHERE Launch_Site LIKE 'CCA%' LIMIT 5;
```

### Tarea 3

##### Muestre la masa de carga útil total transportada por propulsores lanzados por la NASA (CRS)[cite: 3]


```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) AS Total_Payload_Mass FROM SPACEXTABLE WHERE Customer = 'NASA (CRS)';
```

### Tarea 4

##### Muestre la masa de carga útil promedio transportada por la versión del propulsor F9 v1.1[cite: 3]


```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) AS Average_Payload_Mass FROM SPACEXTABLE WHERE Booster_Version = 'F9 v1.1';
```

### Tarea 5

##### Enumere la fecha en que se logró el primer resultado de aterrizaje exitoso en una plataforma terrestre (ground pad).[cite: 3]


_Pista: Use la función min_[cite: 3]


```python
%sql SELECT MIN(Date) AS First_Successful_Landing FROM SPACEXTABLE WHERE "Landing_Outcome" = 'Success (ground pad)';
```

### Tarea 6

##### Enumere los nombres de los propulsores que tienen éxito en una nave no tripulada (drone ship) y tienen una masa de carga útil mayor a 4000 pero menor a 6000[cite: 3]


```python
%sql SELECT Booster_Version FROM SPACEXTABLE WHERE "Landing_Outcome" = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000;
```

### Tarea 7

##### Enumere el número total de resultados de misiones exitosas y fallidas[cite: 3]


```python
%sql SELECT Mission_Outcome, COUNT(*) AS Total_Number FROM SPACEXTABLE GROUP BY Mission_Outcome;
```

### Tarea 8

##### Enumere todas las booster_versions que han transportado la masa de carga útil máxima, utilizando una subconsulta con una función agregada adecuada.[cite: 3]


```python
%sql SELECT Booster_Version FROM SPACEXTABLE WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE);
```

### Tarea 9

##### Enumere los registros que mostrarán los nombres de los meses, los landing_outcomes fallidos en una nave no tripulada (drone ship), las versiones de propulsores, el launch_site para los meses del año 2015.[cite: 3]

**Nota: SQLLite no admite nombres de meses. Por lo tanto, debe usar substr(Date, 6,2) como mes para obtener los meses y substr(Date, 1,4)='2015' para el año.**[cite: 3]


```python
%sql SELECT substr(Date, 6, 2) AS Month, "Landing_Outcome", Booster_Version, Launch_Site FROM SPACEXTABLE WHERE substr(Date, 1, 4) = '2015' AND "Landing_Outcome" = 'Failure (drone ship)';
```

### Tarea 10

##### Clasifique (rank) el recuento de landing_outcomes (como Failure (drone ship) o Success (ground pad)) entre las fechas 2010-06-04 y 2017-03-20, en orden descendente.[cite: 3]


```python
%sql SELECT "Landing_Outcome", COUNT(*) AS Outcome_Count FROM SPACEXTABLE WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY "Landing_Outcome" ORDER BY Outcome_Count DESC;
```
